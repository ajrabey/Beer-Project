```python
import pandas as pd
rev = pd.read_csv('desktop/DATASETS/beer_reviews.csv')
print(rev.shape)
print(rev.head())
print(rev.review_overall.dtype)
```

```python
print(rev.beer_style.value_counts()) #104 beer styles
```

```python
#Comparing Style to Overall review and creating a sorted table
Rev_style_table_mean = rev.groupby('beer_style').review_overall.mean()
Rev_style_table_mean = Rev_style_table_mean.sort_values()
print(Rev_style_table_mean)
#print(Rev_style_table_mean.loc[:, [ >= 3.0]])
```

```python
#Need to sort the two columns in order so we can see the best/worst, plot on Horizontal bar
import matplotlib.pyplot as plt
Rev_style_table_mean.sort_values().plot(kind='barh', stacked=True)
plt.ylabel("Beer Style")
plt.xlabel("Overall Review")
plt.title("Best and Worst Beer Styles")
```

```python
Rev_style_table_freq = pd.crosstab(rev.beer_style, rev.review_overall)
print(Rev_style_table_freq.sort_values('beer_style', ascending=True)) #better visualization of frequency of ratings by each style
```

```python
import matplotlib.pyplot as plt
Rev_style_table_freq = pd.crosstab(rev.beer_style, rev.review_overall)
Rev_style_table_freq.plot(kind='bar', stacked=True) # is actually useless because different beers have more ratings and others have less therefore you need to calculate mean
```

```python
#Create a dataframe of top five best and worst styles for a better plot
import numpy as np
rev['review_overall_int'] = rev['review_overall'].astype(int) # Changed overall reviews to integ
New_Rev_table_mean = rev.groupby('beer_style').review_overall_int.mean() # created subset of beer styles and mean of their reviews
New_Rev_table_mean = New_Rev_table_mean.sort_values(ascending=True) # sorted 
Worst_Styles = New_Rev_table_mean.iloc[:5] #use index loc to get top/bot 5
Best_Styles = New_Rev_table_mean.iloc[-5:]
Best_Worst_Styles = pd.concat([Best_Styles, Worst_Styles])
Best_Worst_Styles.plot(kind='barh',subplots=True)
```
